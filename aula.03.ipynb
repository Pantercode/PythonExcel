```python
#formatar floats
df_tempo = pd.DataFrame(np.random.rand(10,2)*5,
                        index = pd.date_range(start='2023-01-01',periods=10),
                        columns=["Porto Alegre","São Paulo"])
```

```python
df_tempo = df_tempo.round(3)  # Reduz a precisão para 3 casas decimais
df_tempo = df_tempo.applymap(lambda x: str(x).replace('.', ','))  # Substitui ponto por vírgula
```

```python
df_tempo_u = df_tempo.apply(lambda row: row.str.upper(), axis=1)
```

```python
df_tempo_u.style.format_index (lambda row: row.strftime('%d/%m/%Y'), axis=0)
```

```python
#Formatando o back ground
df_tempo_u.style.background_gradient(axis=1, vmin=1, vmax=5, cmap='YlGnBu')
```
