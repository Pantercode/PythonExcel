```python
import pandas as pd
```

```python
#pivotando tabelas
# Lendo os datasets
df_vendas = pd.read_csv('datasets/vendas.csv',sep=';' , decimal=',')

# Lendo os datasets produtos
df_produtos = pd.read_csv('datasets/produtos.csv', sep=';', decimal=',',index_col=0)
```

```python
df_produto = df_produtos.rename(columns={'nome':'produto'})
```

```python
#Merge
df_vendas_uniao = pd.merge(left= df_vendas, 
                           right=df_produto,
                            on= 'produto',
                            how='left')
df_vendas_uniao
```

```python
df_vendas = df_vendas_uniao
df_produtos = df_produto
```

```python
#TABELA DINAMICA
df_vendas.pivot_table(values='preco',index='filial',columns='produto',aggfunc='count')
```

```python
#TABELA DINAMICA
df_vendas.pivot_table(values='preco',index='vendedor',columns='produto',aggfunc='count')
```
